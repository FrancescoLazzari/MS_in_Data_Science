To facilitate the population of our database, we have automatically generated `INSERT TABLE` statements for each record in the database, appropriately saved in `SQL` files. Due to the high number of total records (approximately $350K$), we noticed that not all computers were able to open a single file with all the records for all tables (a much more convenient solution). 

Therefore, we decided to split this process into two separate files. Following this decision, we put the records of the first 4 tables into a first .sql file named `HW1_insert1`, and all the records of the main table `FLIGHTS` into a second file named `HW1_insert2`. Naturally, we are aware that the weight of the two files is significantly unbalanced in favor of the second one, but we preferred to keep all records of a table in the same script.

```python
# Populate HW1_insert1.sql with records from the first 4 tables
with open('HW1_insert1.sql', 'w') as sqlfile:
    # Insert records for Airlines table
    with open('airlines.csv', 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        # Loop through each row of the CSV file
        for row in reader:
            # here we sobstitute all the Nan or NA value into the .csv files with a 'NULL' statement 
            for key in row:
                if row[key] == '':
                    row[key] = 'NULL'
                else:
                    row[key] = '"' + row[key] + '"' # Add quotes for non-empty values
            
            # Construction of the SQL INSERT statement for the current row
            sql_insert = "INSERT INTO DM_HW1.airlines (carrier, name) VALUES ({}, {});\n".format(
                row['carrier'], row['name']
            )
            # Writing of the SQL INSERT statement to the .sql file
            sqlfile.write(sql_insert)

    # Insert records for Airports table
    with open('airports_refined.csv', 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        # Loop through each row of the CSV file
        for row in reader:
            # here we sobstitute all the Nan or NA value into the .csv files with a 'NULL' statement 
            for key in row:
                if row[key] == '':
                    row[key] = 'NULL'
                else:
                    row[key] = '"' + row[key] + '"' # Add quotes for non-empty values

            # Construction of the SQL INSERT statement for the current row
            sql_insert = "INSERT INTO DM_HW1.airports (faa, name, time_zone, tz, dst, latitude, longitude, alt) VALUES ({}, {}, {}, {}, {}, {}, {}, {});\n".format(
                row['faa'], row['name'], row['tzone'], row['tz'], row['dst'], row['lat'], row['lon'], row['alt']
            )
            # Writing of the SQL INSERT statement to the .sql file
            sqlfile.write(sql_insert)

    # Insert records for Planes table
    with open('planes_refined.csv', 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        # Loop through each row of the CSV file
        for row in reader:
            # here we sobstitute all the Nan or NA value into the .csv files with a 'NULL' statement 
            for key in row:
                if row[key] == '':
                    row[key] = 'NULL'
                else:
                    row[key] = '"' + row[key] + '"' # Add quotes for non-empty values

            # Construction of the SQL INSERT statement for the current row
            sql_insert = "INSERT INTO DM_HW1.planes (tail_num, year, manufacturer, model, type, n_engines, seats, speed, engine_type, disused) VALUES ({}, {}, {}, {}, {}, {}, {}, {}, {}, {});\n".format(
                row['tailnum'], row['year'], row['manufacturer'], row['model'], row['type'], row['engines'], row['seats'], row['speed'], row['engine'], row['disused']
            )
            # Writing of the SQL INSERT statement to the .sql file
            sqlfile.write(sql_insert)

    # Insert records for Weather table
    with open('weathers_refined.csv', 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        # Loop through each row of the CSV file
        for row in reader:
            # here we sobstitute all the Nan or NA value into the .csv files with a 'NULL' statement 
            for key in row:
                if row[key] == '':
                    row[key] = 'NULL'
                else:
                    row[key] = '"' + row[key] + '"' # Add quotes for non-empty values

            # Construction of the SQL INSERT statement for the current row
            sql_insert = "INSERT INTO DM_HW1.weather (date_hour, airports_origin, temp, humid, precipitation, wind_speed, wind_dir, visibility, dewp, pressure, wind_gust) VALUES ({}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {});\n".format(
                row['time_hour'], row['origin'], row['temp'], row['humid'], row['precip'], row['wind_speed'], row['wind_dir'], row['visib'], row['dewp'], row['pressure'], row['wind_gust']
            )
            # Writing of the SQL INSERT statement to the .sql file
            sqlfile.write(sql_insert)

# -------------------------------------------------------------------------

# Populate HW1_insert2.sql with records from the Flights table
with open('HW1_insert2.sql', 'w') as sqlfile:
    # Insert records for Flights table
    with open('flights_refined.csv', 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        # Loop through each row of the CSV file
        for row in reader:
            # here we sobstitute all the Nan or NA value into the .csv files with a 'NULL' statement 
            for key in row:
                if row[key] == '':
                    row[key] = 'NULL'
                else:
                    row[key] = '"' + row[key] + '"' # Add quotes for non-empty values
            
            # Construction of the SQL INSERT statement for the current row
            sql_insert = "INSERT INTO DM_HW1.flights (flight_ID, carrier, tail_num, date_hour, origin, destination, sched_dep, dep_delay_min, sched_arrive, arr_delay_min, distance, air_time_min, canceled) VALUES ({}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {}, {});\n".format(
                row['flight'], row['carrier'], row['tailnum'], row['time_hour'], row['origin'], row['dest'], row['sched_dep_time'], row['dep_delay_min'], row['sched_arr_time'], row['arr_delay_min'], row['distance'], row['air_time_min'], row['canceled']
            )
            # Writing of the SQL INSERT statement to the .sql file
            sqlfile.write(sql_insert)
```
